```python
# CARD
# SUIT (Hearts, Clubs, Diamonds, Spades), 
# RANK (Ace, 2, 3, .... , Jack, Queen, King), 
# VALUE (An actual integer value set to a rank defining it's priority)
from IPython.display import clear_output
import random
import math
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}
```

```python
class Card:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                # Create the card object
                created_card = Card(suit, rank)
                self.all_cards.append(created_card)
    
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player:
    
    def __init__(self, name):
        
        self.name = name
        self.all_cards = []
        self.val = 0
        self.chips = 100
        
    def hit(self, new_cards):
        self.all_cards.append(new_cards)
        
        self.val += new_cards.value
        
        if new_cards.rank == 'Ace':
            if self.val + 10 <= 21:
                self.val += 10 
    
    def display(self, show = True):
        print(f"\n{self.name}'s cards:")
        if not show:
            for i in range(len(self.all_cards)):
                if i == 0:  
                    print("Flipped Card")
                else:
                    print(self.all_cards[i])
        
        else:
            for crds in self.all_cards:
                print(crds)
    
    def BlackJack(self):
        return (self.val == 21 and len(self.all_cards) == 2)
    
    def Bust(self):
        return(self.val > 21)
            
    
```

```python
def continue_game():
    x = ''
    while x not in ('y','n','Y','N'):
        x = input("play again (y/n): ")
        if x not in ('y','n','Y','N'):
            print("Invalid Input")
    if x == 'y':
        clear_output()
        return True
    else:
        return False
```

```python
def decorator(Title_func):
    print("***********************************************")
    
    print("\n")
    
    Title_func()
    
    print("\n")
    
    print("***********************************************")
    
    print("\n")
```

```python
game_on = True
while game_on:
    
    Hitting = True
    player = Player("Player one")
    dealer = Player("Dealer")
    bet = ''
    
    decorator(Title)
    
    while bet not in range(0,101):
        try:
            bet = int(input("Enter betting amount (0-100):"))
        except:
            print("Invalid input")
        else:
            if bet not in range(0,101):
                print("Invalid input")
        
        
    new_deck = Deck()
    new_deck.shuffle()

    #test case
    #player.hit(Card('Hearts', 'Ace'))
    #player.hit(Card('Spades', 'King'))
    #dealer.hit(Card('Hearts', 'Queen'))
    #dealer.hit(Card('Spades', 'King')) 
    
    for x in range(2):
        player.hit(new_deck.deal_one())
        dealer.hit(new_deck.deal_one())
    
    dealer.display(False)
    player.display()
    
    if player.BlackJack():
        if dealer.BlackJack():
            dealer.display()
            print(f"Dealer has value: {dealer.val}")
            player.display()
            print(f"Player has value: {player.val}")
            print("Both hits a Black Jack! Game Push!")
            print(f"Player's winnings stand at: {player.chips}")
            
        
        else:
            dealer.display()
            print(f"Dealer has value: {dealer.val}")
            player.display()
            print(f"Player has value: {player.val}")
            print("You hit a Black Jack! You win!")
            player.chips += math.ceil(bet*1.5)
            print(f"Player's winnings stand at: {player.chips}")
            
            
    else:
        
        #Player's turn:
        while Hitting:
            choice = ''
            while choice not in (1,2):
                try:
                    choice = int(input("Press 1 to hit, 2 to stay: "))
                except:
                    print("Invalid input")
                else:
                    if choice not in (1,2):
                        print("Invalid input")
                    
            if choice == 1:
                player.hit(new_deck.deal_one())
                dealer.display(False)
                player.display()
                if player.Bust():
                    dealer.display()
                    print(f"Dealer has value: {dealer.val}")
                    player.display()
                    print(f"Player has value: {player.val}")
                    print("You Bust! Dealer Wins")
                    player.chips -= bet
                    print(f"Player's winnings stand at: {player.chips}")
                    Hitting = False
                
                elif player.val == 21:
                    Hitting = False  
                
            
            else:
                Hitting = False
        
        
        
        #Dealer's Turn
        
        while dealer.val < 17 and player.Bust() != True and player.BlackJack() != True:
            dealer.hit(new_deck.deal_one())
          
        
        if dealer.Bust():
            dealer.display()
            print(f"Dealer has value: {dealer.val}")
            player.display()
            print(f"Player has value: {player.val}")
            print("Dealer Bust! You Win")
            player.chips += bet
            print(f"Player's winnings stand at: {player.chips}")
            
            
        elif dealer.BlackJack() and not player.Bust():
            dealer.display()
            print(f"Dealer has value: {dealer.val}")
            player.display()
            print(f"Player has value: {player.val}")
            print("Dealer hits Blackjack! Dealer Wins")
            player.chips -= bet
            print(f"Player's winnings stand at: {player.chips}")
            
        elif dealer.Bust() != True and player.Bust()!= True and player.BlackJack() != True and dealer.BlackJack() != True:
            
            dealer.display()
            print(f"Dealer has value: {dealer.val}")
            
            player.display()
            print(f"Player has value: {player.val}")
            
            if dealer.val > player.val:
                print("Dealer Wins!")
                player.chips -= bet
                print(f"Player's winnings stand at: {player.chips}")
            
            elif dealer.val < player.val:
                print("Player Wins!")
                player.chips += bet
                print(f"Player's winnings stand at: {player.chips}")
            
            else:
                print("Game Push")
                print(f"Player's winnings stand at: {player.chips}")
    
    game_on = continue_game()
        
        
        

        
            
    
    
```
